```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
import json
import os
from tqdm import tqdm
from lxml import etree
import ssl 

ssl._create_default_https_context = ssl._create_unverified_context 
```

```python
content = urlopen('https://tei-c.org/Vault/MembersMeetings/2012/teiconference/program/papers/#mads')
HTML = content.read()

if not os.path.exists("./cache2012"):
    os.makedirs("./cache2012")
if not os.path.exists("./cache2012/cacheXML"):
    os.makedirs("./cache2012/cacheXML")
Path = './cache2012/cacheXML/'

soup = BeautifulSoup(HTML, 'html.parser')
liste_des_abstracts = {}
i = 1
for hr in soup.find_all("hr"):
    with open(Path + str(i), 'w', encoding='UTF-8') as y:
        for item in hr.find_next_siblings():
            y.write(str(item))
            if item.name == "hr":
                break
    i += 1

j = 0
dict = {}
list = []
for p in soup.find_all('span', style="text-decoration: underline"):
    if '@' not in str(p):
        list.append(str(p))
    else:
        dict[j] = list
        j += 1
        list = []
```
